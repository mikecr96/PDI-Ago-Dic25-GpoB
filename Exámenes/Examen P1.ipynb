## Parte 3: Transformaciones de Intensidad (25 puntos)

### Ejercicio 3.1 - Corrección Gamma (9 puntos)
Implemente la transformación gamma s = c * r^γ, donde c es una constante de normalización.
Cree una función que permita ajustar el valor de gamma y muestre los resultados para γ = 0.3, γ = 1.0, y γ = 2.5.

```python
def correccion_gamma(imagen, gamma):
    """
    Aplica corrección gamma a una imagen
    """
    # Su código aquí
    
    return imagen ** gamma

# Aplicar corrección gamma con diferentes valores
gammas = [0.3, 1.0, 2.5]

# Su código para mostrar los resultados
lista = []
for g in gammas:
    lista.append(correccion_gamma(img_gray, g))

for contador, img in enumerate(lista):
    plt.imshow(img, cmap="gray")
    plt.title(f"Transformación con gamma={gammas[contador]}")
    plt.show()
```

### Ejercicio 3.2 - Transformación Negativo (8 puntos)
Implemente la transformación negativo s = L - 1 - r, donde L=256 para imágenes de 8 bits.
Aplique esta transformación a la imagen en escala de grises y muestre la imagen original y el negativo lado a lado.

```python
def transformacion_negativo(imagen):
    """
    Aplica transformación negativo a una imagen
    """
    # Su código aquí
    
    return 256-1-imagen

# Aplicar transformación y mostrar resultado
img_gray_negativo = transformacion_negativo(img_gray)
plt.imshow(np.hstack((img_gray, img_gray_negativo)), cmap="gray")
plt.show()
```

### Ejercicio 3.3 - Transformación Logarítmica (8 puntos)
Implemente la transformación logarítmica s = c * log(1 + r), donde c es una constante.
Use c = 255 / np.log(1 + np.max(imagen)). Compare la imagen original con la transformada mostrando ambas.

```python
def transformacion_logaritmica(imagen):
    """
    Aplica transformación logarítmica a una imagen
    """
    # Su código aquí
    c = 255 / np.log10(1 + np.max(imagen).item())
    s = c * np.log10(1 + imagen)
    
    return s

# Aplicar transformación y mostrar resultado
img_gray_log = transformacion_logaritmica(img_gray)
plt.imshow(img_gray_log, cmap="gray")
plt.show()
```
