## Transformación de umbralización

```python
''''
k = 127
si r < k : s = 0
si r >= k : s = 255
'''

# logo_diablos_negative = logo_diablos_gray.copy()
logo_diablos_umbral = np.zeros(logo_diablos_gray.shape, dtype=np.uint8)
k = 20

for x in range(num_filas):
    for y in range(num_columnas):
        if logo_diablos_gray[x, y] >= k:
            logo_diablos_umbral[x, y] = 255
        else:
            logo_diablos_umbral[x, y] = 0
```

```python
plt.imshow(logo_diablos_umbral, cmap='gray')
plt.title('Imagen en blanco y negro')
plt.show()
```

```python
# logo_diablos_negative = logo_diablos_gray.copy()
logo_diablos_log = np.zeros(logo_diablos_gray.shape, dtype=np.uint8)

for x in range(num_filas):
    for y in range(num_columnas):
        logo_diablos_log[x, y] = np.log10(logo_diablos_gray[x, y] + 1)
```
