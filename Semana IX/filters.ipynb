```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernel_caja = np.ones((3, 3), dtype=np.uint8)
```

```python
apple_filterbox = cv.filter2D(apple, -1, kernel_caja)
```

```python
plt.imshow(apple_filterbox, cmap="gray")
plt.show()
```

```python
kernel_caja
```

```python
(1/9) * kernel_caja
```

```python
kernel_caja_normalizado = (1/9) * kernel_caja
```

```python
apple_filterbox_norm = cv.filter2D(apple, -1, kernel_caja_normalizado)
```

```python
plt.imshow(np.hstack((apple, apple_filterbox_norm)), cmap="gray")
plt.show()
```

```python
np.ones((5, 5)) * 1/5**2
```

```python
def crearKernelCajaNormalizado(size:int):
    return np.ones((size, size)) * 1/size**2
```

```python
crearKernelCajaNormalizado(3)
```

```python
crearKernelCajaNormalizado(4)
```
