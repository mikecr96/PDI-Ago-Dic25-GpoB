```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
mariposa_g = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic25/Grupo B/images/butterfly.jpg", 0)
```

```python
def showImg(img, title=""):
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.show()
```

```python
showImg(mariposa_g)
```

---
## Umbralizacin Global

```python
# Con OpenCV
showImg(cv.threshold(mariposa_g, 40, 255, cv.THRESH_BINARY_INV)[1])
```

```python
# Con NumPy
showImg(np.where(mariposa_g >= 40, 0, 255))
```

```python
umbral, mariposa_b = cv.threshold(mariposa_g, 10000, 255, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)
```

```python
showImg(mariposa_b, umbral)
```

```python
umbral, mariposa_b = cv.threshold(mariposa_g, 0, 255, cv.THRESH_BINARY_INV | cv.THRESH_TRIANGLE)
showImg(mariposa_b, umbral)
```

```python
sudoku_g = cv.imread("/Users/miguelcamargorojas/Documents/UP/PDI-Ago-dic25/Grupo B/images/sudoku.png", 0)
```

```python
showImg(sudoku_g)
```

```python
umbral, mariposa_b = cv.threshold(sudoku_g, 0, 255, cv.THRESH_BINARY | cv.THRESH_OTSU)
showImg(mariposa_b, umbral)
```

```python
umbral, mariposa_b = cv.threshold(sudoku_g, 0, 255, cv.THRESH_BINARY | cv.THRESH_TRIANGLE)
showImg(mariposa_b, umbral)
```

```python
umbral, mariposa_b = cv.threshold(sudoku_g, 45, 255, cv.THRESH_BINARY)
showImg(mariposa_b, umbral)
```

## Umbralizacin Adaptativa

```python
showImg(cv.adaptiveThreshold(sudoku_g, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 21, 7.5))
```

```python
showImg(cv.adaptiveThreshold(sudoku_g, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2))
```
